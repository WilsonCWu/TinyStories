```python
from transformers import AutoModelForCausalLM, AutoTokenizer, GenerationConfig
tokenizer = AutoTokenizer.from_pretrained("EleutherAI/gpt-neo-125M")

model_ts_1m = AutoModelForCausalLM.from_pretrained('roneneldan/TinyStories-1M')
model_ts_33m = AutoModelForCausalLM.from_pretrained('roneneldan/TinyStories-33M')
model_tsi_33m = AutoModelForCausalLM.from_pretrained('roneneldan/TinyStories-Instruct-33M')
```

```python
prompt = "Once upon a time there was a baby in a"
input_ids = tokenizer.encode(prompt, return_tensors="pt")
attention_mask = torch.ones_like(input_ids)
pad_token_id = tokenizer.eos_token_id
output = model_ts_1m.generate(input_ids, max_length = 1000, num_beams=5, attention_mask=attention_mask, pad_token_id=pad_token_id)
output_text = tokenizer.decode(output[0], skip_special_tokens=True)
print(output_text)
```
